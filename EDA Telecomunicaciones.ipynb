Borro las filas nulas ya que representan poco del total de los datos.

```python
data_four_clear = data_four.dropna()
```

```python
data_four_clear.isnull().sum()
```

```python
data_four_clear["Año"] = data_four_clear["Año"].apply(int)
data_four_clear["Trimestre"] = data_four_clear["Trimestre"].apply(int)
```

```python
data_four_clear
```

```python
data_four_clear["Banda_ancha_fija"] = data_four_clear["Banda_ancha_fija"].apply(str)
data_four_clear["Dial_up"] = data_four_clear["Dial_up"].apply(str)
data_four_clear["Total"] = data_four_clear["Total"].apply(str)
```

```python
data_four_clear["Banda_ancha_fija"] = data_four_clear["Banda_ancha_fija"].str.replace(".", "")


data_four_clear["Dial_up"] = data_four_clear["Dial_up"].str.replace(".", "")


data_four_clear["Total"] = data_four_clear["Total"].str.replace(".", "")
```

```python
data_four_clear["Banda_ancha_fija"] = data_four_clear["Banda_ancha_fija"].astype("float")
```

```python
data_four_clear["Total"] = data_four_clear["Total"].astype("float")
```

```python
data_four_clear["Dial_up"] = pd.to_numeric(data_four_clear["Dial_up"], errors='coerce')
```

```python
data_four_clear["Dial_up"] = data_four_clear["Dial_up"].astype("float")
```

```python
data_four_clear
```

```python
data_four_clear.info()
```

Reemplazo los Nan en la columna Dial_up por la media de todos los datos de la columna

```python
mean = data_four_clear["Dial_up"].mean()
data_four_clear["Dial_up"].fillna(mean, inplace=True)
```
